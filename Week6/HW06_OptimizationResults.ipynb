Increasing the learning rate has resulted in not being able to finding the optimum. We decided to keep the learning rate at 0.001.

For our next step in optimization, we introduce a dropout layer. We place this layer between the second and third block of our residual network. After 2 blocks, the model has a dropout layer and then it will go through 2 more blocks until the output layer.
