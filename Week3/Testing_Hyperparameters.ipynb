# Takeaway

- When we decreased the batch size from 32 to 8, it affected the computation time immensely. It doubled the time to train the model.
- We could not observe much difference when we added momentum to SGD. 
- Decrease in unit sizes of the dense layers (256 to 64) also did not show much difference. Further tests with 16 units in each layer had slight difference where the training loss and test loss have some disparity. However, only slightly.
- Decrease in epoch size and the learning rate did not make difference again.
- We interpret these results as that the MNIST dataset is easy to train on. The task is not difficult, therefore a model that is not strong can be trained to perform well enough for this task.
